```python
import pandas as pd
import string
import numpy as np
```

```python
# 讀取看門狗的檔案
data = pd.read_excel('/Users/wangyuda/Desktop/文字探勘/Model/all.xlsx')
```

```python
# 引入操作WordPress所需要的套件
from wordpress_xmlrpc import Client, WordPressPost
from wordpress_xmlrpc.methods.users import GetUserInfo
from wordpress_xmlrpc.methods.posts import GetPosts, NewPost

# 網站登入資訊，帳號與密碼
id = "p314008dsa1"
password = "s7039970399"

# 輸入我們的WordPress網址，若試著連上該網址，應該會出現「XML-RPC server accepts POST requests only.」才對。
url = "https://pa566554685.wordpress.com/xmlrpc.php"

# 新文章要直接發布的話，就不用改，如果要變成草稿，就改成"draft"
which = "publish"
#which="draft"

# 建立客戶端，登入網頁
wp = Client(url, id, password)
```

```python
# 先試試前十筆資料的上傳狀況
for i in range(1, 10):
    title = str(data.iloc[i, 2]) + ' ' + str(data.iloc[i, 1])                             # 文章標題
    content = data.iloc[i, 7]                                                               # 文章內容
    excerpt = "事件強度：" + str(data.iloc[i, 4]) + '\n' + "預期漲跌幅："   # 文章摘要
    category = str(data.iloc[i, 0]) + ' ' + str(data.iloc[i, 1])                      # 文章分類
    # post_tag =''                                                                            # 文章標籤
    
    # 依據讀取的資料，在部落格中建立新文章
    post = WordPressPost()
    post.post_status = which
    post.excerpt = excerpt
    post.title = title
    post.content = content
    
    # 新增文章標籤或分類
    # 標籤可以給關鍵字
    post.terms_names = {
    # "post_tag": ["標籤一","標籤二"],
    "category": [category]
    }

    # 如果這一篇是過去的文章，可以透過這個方式指定該文章發表的日期。
    # post.date=datetime.strptime("2018/1/01 10:05:10","%Y/%m/%d %H:%M:%S")
    
    # 要發布的資料都輸入完畢後，發布文章
    wp.call(NewPost(post))
```

```python
import requests
import json
import csv
from bs4 import BeautifulSoup
import pandas as pd
import time,datetime,os
```

```python
dt=datetime.datetime.now()
print(dt.year)
print(dt.month)
print(dt.day)
date=str(dt.year)+"{0:0=2d}".format(dt.month)+str(dt.day)
```

```python
stocknumber='2880'
```

```python
url='http://www.twse.com.tw/exchangeReport/STOCK_DAY?response=json&date='+date+'&stockNo='+stocknumber
```

```python
doc = requests.post(url,)
soup = BeautifulSoup(doc.text, "html.parser")
newdoc=json.loads(soup.text)
newdoc
```

```python
stockdata=pd.DataFrame(newdoc['data'],columns=newdoc['fields'])
stockdata
```

```python
data=pd.read_csv('ckiptagger_data.csv')
```

```python
df=data.dropna(inplace=False)
```

```python
TCRI=[]
for i in range(0,len(df)):
    TCRI.append(df.iloc[i,3][0])
```

```python
df=df.drop(columns=['TCRI(年/月)'])
len(TCRI)
```

```python
df.insert(3, 'TCRI', TCRI)
```

```python
words=[]
for word in df.content_tokenized.iloc[:100]:
    a=word.split(" ")
    for b in a:
        words.append(b)
```

```python
data=[]
onematrix=[]
```

```python
import math
for a in df.iloc[0:100,7]:
    for b in words:
        count = 0
        if (a.find(b)>=0):
            count += 1
        onematrix.append(count)
    data.append(onematrix)
    onematrix=[]
```

```python
data1=pd.DataFrame(data,index=df.iloc[0:100,7],columns=words)
data1
```

```python
sns.heatmap(data1)
```

```python
cooccurrence_matrix = np.dot(data1.transpose(),data1)
```

```python
data2=pd.DataFrame(cooccurrence_matrix,index=words,columns=words)
data2
```

```python
sns.heatmap(data2)
```

```python
co_mat = []
row = data2.iloc[:,2]
row
for j in range(0,len(data2)):
    if row[j] != 0:
        co_mat.append(('調動',words[j],row[j]*0.001))
```

```python
import networkx as nx
import numbers
g = nx.Graph()
g.add_weighted_edges_from(co_mat)
pos=nx.spring_layout(g)
nx.draw_networkx_nodes(g, pos, alpha=0.5,node_size=[val*60 for (node, val) in g.degree()], node_color='#08c5b1')
nx.draw_networkx_edges(g, pos, node_color='b', alpha=0.2)
nx.draw_networkx_labels(g, pos, font_size = 10, font_family='sans-serif', alpha=0.8)
plt.title('調動的共線圖', size = 10)
plt.show()
```
